### VIX (CBOE Volatility Index) is a popular measure of the stock market's expectation of volatility implied by S&P 500 index options. This data will be used to determine if volatility in the markets impacts price of bitcoin. 

```python
#Import required packages 
import numpy as np
import pandas as pd
import pandas_profiling
from datetime import datetime
import pickle
```

```python
vix_volatility_hourly_history_df = pd.read_csv('./data/vix_volatility_hourly_history.csv')
```

```python
vix_volatility_hourly_history_df['date'] = vix_volatility_hourly_history_df['Local time'].astype(str).str[5:]
```

```python
vix_volatility_hourly_history_df['date_1'] = vix_volatility_hourly_history_df['Local time'].astype(str).str[0:2]
```

```python
vix_volatility_hourly_history_df['date_2'] = vix_volatility_hourly_history_df['Local time'].astype(str).str[3:5]
```

```python
vix_volatility_hourly_history_df['local_time'] = vix_volatility_hourly_history_df['date_2'] + '.' + vix_volatility_hourly_history_df['date_1'] + vix_volatility_hourly_history_df['date']
vix_volatility_hourly_history_df.drop(['date', 'date_1', 'date_2'], axis = 1, inplace = True)
```
